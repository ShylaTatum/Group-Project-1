## Do Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', None)
```

# Import the cleaned up WHO data file to visualize.

```python
# Import the CSV
df_final = pd.read_csv('df_final.csv')
display(df_final.shape)

#Remove rows with ICD Code 'AAA' since it is double counting the death counts.
df_final = df_final[df_final['ICD Code'] != 'AAA']
display(df_final.shape)
```

# Plot the cleaned data to show mortality classification by age group.

```python
# Grouping by Year and Country Name, then summing up the deaths across age groups
mortality_by_year_icd = df_final.groupby(['Year']).agg({'Deaths: 0-1 Years': 'sum', 'Deaths: 1-24 Years': 'sum', 'Deaths: 25-44 Years': 'sum', 'Deaths: 45-69 Years': 'sum', 'Deaths: 70-95+ Years': 'sum'})

#Plot a stacked line plot for mortality data by year across all age groups. 
#Customize the plot with labels, axis limits, and a legend, and configure to display a maximum of five ticks on the X-axis. 
#Set the Y-axis range between 100 and 40,000 deaths, and position the legend outside the plot for clarity.
plot = mortality_by_year_icd.plot(kind='line', stacked=False, figsize=(10, 5), colormap='tab20')
plot.set_xlabel('Year')
plot.set_ylabel('Number of Deaths')
plot.set_title('Mortality by Year and Age Group')
plot.xaxis.set_major_locator(plt.MaxNLocator(5))
plot.legend(loc='upper left', title='Age Group', bbox_to_anchor=(1, 1))
plot.ticklabel_format(style='plain', axis='y')

plt.show()

```

## Mortality by Year and Age Group Analysis Summary
 - The graph indicates that mortality rates increase with age, with the 70-95+ age group consistently experiencing the highest number of deaths.
 - There is a noticeable increase in mortality from 2019 to 2020, especially for the older age groups,  reflecting the impact of COVID-19 pandemic.


```python
#Using pivot table to plot the data for the age group All Ages and ICD category for each year from 2017 to 2021.
df_final_pivot = df_final.pivot_table(index='Year', columns='ICD Category', values='Deaths: All Ages', aggfunc='sum')

#Plot a stacked line plot for mortality data by year, with a focus on the age group All Ages. 
#Customize the plot with labels, axis limits, and a legend, and configure to display a maximum of five ticks on the X-axis. 
#Set the Y-axis range between 100 and 40,000 deaths, and position the legend outside the plot for clarity.
plot = df_final_pivot.plot(kind='line', stacked=True, figsize=(10, 5), colormap='tab20')
plot.set_xlabel('Year')
plot.set_ylabel('Number of Deaths')
plot.xaxis.set_major_locator(plt.MaxNLocator(5))
plot.set_ylim(2000000, 9000000)
plot.set_title('Mortality by Year and ICD Category')
plot.legend(loc='upper left', title='Age Group - All Ages', bbox_to_anchor=(1, 1))
plot.ticklabel_format(style='plain', axis='y')

plt.show()
```

## Mortality by Year and ICD Category for Age Groups Analysis Summary
- Respiratory System Diseases, Circulatory System Diseases, and Neoplasms have the highest mortality rates, with noticeable increas particularly after 2019.
- Infectious and Parasitic Diseases show a significant increase in mortality around 2020 due to the COVID-19 pandemic.
- The lowest mortality rates are associated with categories like Pregnancy, Childbirth and the Puerperium and Congenital Malformations.
