## Basic Cleaning Findings

**Null Columns** |
The **Admin1** column is always null. In this dataset, it has no values. Recommendation: We should drop this column in our cleaned DataFrame.

**Missing Information** |
The `country_codes`, `subdiv_codes`, and `ICD10`,  CSV files in the `Resources` folder contain essential information about the data in the morbidity dataset. Recommendation: We should add columns to the cleaned DataFrame by merging in the other CSVs.
* Country name (country_codes CSV)
* Subdivision description (subdiv_codes CSV)
* ICD 10 code description (icd10_codes_who CSV)

**Null Values in Rows** |
Some rows have null values, which show up as `NaN` in the DataFrame. Recommendation: We preserve rows that have null values since the remaining data in that row is essential. However, we should fill that empty cell with a value like `-`. 

**Column Names and Organization** | 
After merging in other datasets, we may have duplicate columns and columns with irregular names. The morbidity dataset includes column names age ranges that aren't indicataive of the actual ages. We have discussed renaming these to be the actual age buckets instead. Recommendation: Drop duplicate columns after merging data and rename columns for clarity in the cleaned DataFrame.

```python
# Drop null 'Admin1' column from the dataframe
df_mort = df_mort.drop(columns=['Admin1'])
```

```python
## The morbidity dataset has the country code in it, but not the country name. Let's merge in df_countries to get the country name.

# Merge df_mort with df_countries on the country code columns
df_mort = df_mort.merge(df_countries[['country', 'name']], left_on='Country', right_on='country', how='left')

# Drop the redundant 'country' column from the merged dataframe
df_mort = df_mort.drop(columns=['country'])
```

```python
# Add the Subdivision Description column
df_mort = df_mort.merge(df_subdiv[['SubDiv', 'SubDiv_Desc']], how='left')
```

```python
# Rename columns
df_mort = df_mort.rename(columns={'name': 'Country Name', 
                                  'Country': 'Country Code', 
                                  'Cause' : 'ICD Code'
                                  })
```

```python
df_mort.head()
```

```python
# Add the ICD 10 code description as a new column
# Merge df_mort with df_icd_codes on the 'ICD Code' column
df_icd10 = df_icd10.rename(columns={'code': 'ICD Code', 'detailed_list_numbers': 'Detailed Codes', 'cause': 'ICD Detail'})

# Convert ICD Code columns to string type if necessary
df_mort['ICD Code'] = df_mort['ICD Code'].astype(str)
df_icd10['ICD Code'] = df_icd10['ICD Code'].astype(str)

# Merge df_mort with df_icd10 on the 'ICD Code' column
df_mort = df_mort.merge(df_icd10, on='ICD Code', how='left')

# Display the first few rows of the updated dataframe
df_mort.head()
```

```python
# Reorganize columns
cols = df_mort.columns.tolist()

# Move 'Country Name' next to 'Country Code'
country_code_index = cols.index('Country Code')
country_name_index = cols.index('Country Name')
cols.insert(country_code_index + 1, cols.pop(country_name_index))

# Move 'Detailed Codes' and 'ICD Detail' next to 'ICD Code'
icd_code_index = cols.index('ICD Code')
detailed_codes_index = cols.index('Detailed Codes')
icd_detail_index = cols.index('ICD Detail')
cols.insert(icd_code_index + 1, cols.pop(detailed_codes_index))
cols.insert(icd_code_index + 2, cols.pop(icd_detail_index))

# Reassign the reordered columns to the DataFrame
df_mort = df_mort[cols]

# Display the first few rows of the updated dataframe
df_mort.head()
```

```python
# Fill in blanks with "-"
df_mort = df_mort.fillna(' - ')
df_mort.head()
```
