```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the WHO mortality data
df_vrovisual = pd.read_csv('Resources/source_data/current_who_mortality_2017_2021.csv')
```

```python
# Plot of All Deaths via Select Countries
# Group by 'Country Name' and 'Year' and sum 'Deaths: All Ages'
df_filter_aaa = df_vrovisual[df_vrovisual['ICD Code'] != 'AAA']
df_grouped = df_filter_aaa.groupby(by=['Country Name', 'Year'])['Deaths: All Ages'].sum().reset_index()
# List of countries to filter
countries_to_include = ['United States of America', 'Brazil', 'Singapore','Iceland']
# Filter the DataFrame based on the list of countries
df_filtered = df_grouped[df_grouped['Country Name'].isin(countries_to_include)]
# Display the filtered DataFrame
display(df_filtered)
# Pivot the DataFrame to have 'Year' as the index and 'Country Name' as columns
df_pivot = df_filtered.pivot(index='Year', columns='Country Name', values='Deaths: All Ages')
# Plot the DataFrame with 'Year' as the x-axis and separate bars for each country
ax = df_pivot.plot(kind='bar', figsize=(18, 9), title='Deaths by Year and Country')
# Add labels
ax.set_xlabel('Year')
ax.set_ylabel('Deaths:')
# Show the plot
plt.show()
```

```python
# Plot for COVID Deaths across Select Countries
# Group by 'Country Name' and 'Year' and sum 'Deaths: All Ages'
df_filter_aaa = df_vrovisual[df_vrovisual['ICD Code'].isin(['U071', 'U072'])]
df_grouped = df_filter_aaa.groupby(by=['Country Name', 'Year'])['Deaths: All Ages'].sum().reset_index()
# List of countries to filter
countries_to_include = ['United States of America', 'Brazil', 'Singapore','Iceland']
# Filter the DataFrame based on the list of countries
df_filtered = df_grouped[df_grouped['Country Name'].isin(countries_to_include)]
# Display the filtered DataFrame
display(df_filtered)
# Pivot the DataFrame to have 'Year' as the index and 'Country Name' as columns
df_pivot = df_filtered.pivot(index='Year', columns='Country Name', values='Deaths: All Ages')
# Plot the DataFrame with 'Year' as the x-axis and separate bars for each country
ax = df_pivot.plot(kind='bar', figsize=(18, 9), title='Deaths Select Country')
# Add labels
ax.set_xlabel('Yrs: 2017 - 2022')
ax.set_ylabel('Deaths:')
# Show the plot
plt.show()
```
